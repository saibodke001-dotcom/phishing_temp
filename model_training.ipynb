# PhishGuard - Explainable AI (XAI) & EDA

This notebook demonstrates the Exploratory Data Analysis (EDA) and SHAP (SHapley Additive exPlanations) visualizations for the PhishGuard Enterprise project.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import shap
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('dataset/phishing_dataset.csv')
df.head()
```

## 1. Interactive EDA (Plotly)

```python
fig = px.pie(df, names='Label', title='Class Distribution (0 = Safe, 1 = Phishing)', hole=0.3)
fig.show()
```

```python
corr = df.corr()
fig = px.imshow(corr, text_auto='.2f', aspect='auto', title='Feature Correlation Heatmap', color_continuous_scale='Viridis')
fig.show()
```

## 2. SHAP Explainability
Load the trained model and explainer to see exactly how the model makes decisions.

```python
X = df.drop('Label', axis=1)
with open('model/phishing_model.pkl', 'rb') as f:
    model = pickle.load(f)

with open('model/shap_explainer.pkl', 'rb') as f:
    explainer = pickle.load(f)

shap_values = explainer.shap_values(X)
if isinstance(shap_values, list):
    shap_values = shap_values[1]

shap.summary_plot(shap_values, X, plot_type='bar')
```
